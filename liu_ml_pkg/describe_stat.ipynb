```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_corr(corr_mat):
# Generate a mask for the upper triangle
    mask = np.zeros_like(corr_mat, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
    f, ax = plt.subplots(figsize=(5, 5))

# Generate a custom diverging colormap
    cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
    sns.heatmap(corr_mat, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
    print(corr_mat)
    plt.show()
plot_corr(data.corr())
```

# Feature engineering

```python
data
time_delay = 10
target='Buying_Rate'
features = [i for i in data.columns if i != target]
```

```python
features
```

```python
data.head(2)
```

```python
rolling_window = 20
```

```python
#create delay variables' last value
X_delay_last_value = data.shift(time_delay)
X_delay_last_value.columns = [i+'_delay' for i in X_delay_last_value.columns]
X_delay_last_value
```

```python
#create moving deviance

X_delay_deviance = data.shift(time_delay).rolling(rolling_window).std()
X_delay_deviance.columns = [i+'_deviance' for i in X_delay_deviance.columns]

X_delay_deviance
```

```python
#create moving deviance

X_delay_pente = data.shift(time_delay).rolling(rolling_window).max() - data.shift(time_delay).rolling(rolling_window).min()
X_delay_pente.columns = [i+'_pente' for i in X_delay_pente.columns]

X_delay_pente
```

# Modelling

```python
print(X_delay_last_value.shape, X_delay_deviance.shape, X_delay_pente.shape)
```

```python
#paste different features
X = pd.concat([X_delay_last_value,X_delay_deviance,X_delay_pente], axis=1)
print(X.shape)
X.head(2)
```

```python
y = data[target]
y.shape
```

```python
X = X[(time_delay+rolling_window):]
y = y[(time_delay+rolling_window):]
```

```python
train_index = np.arange(len(y)) < 0.7 * len(y)
test_index = np.logical_not(train_index)

def score_(pred, obs):
    mae = np.nanmean(abs(pred - obs))
    rmse = np.sqrt(np.nanmean( np.power(pred - obs, 2 )))
    mape = np.nanmean(abs((pred - obs)/obs))
    return mae, rmse, mape
```

```python
#linear model
```

```python
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA 
from sklearn.linear_model import LinearRegression

linear_model = Pipeline([('select_variable', PCA(n_components=3)),
                         ('modelling', LinearRegression())])
linear_model.fit(X[train_index], y[train_index])
linear_pred = linear_model.predict(X[test_index])
linear_mae, linear_rmse, linear_mape = score_(linear_pred, y[test_index].values)
linear_mae, linear_rmse, linear_mape
```

```python
#random forest and grandiant boosting
```

```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100)
rf.fit(X[train_index], y[train_index])
rf_pred = rf.predict(X[test_index])
rf_mae, rf_rmse, rf_mape = score_(rf_pred, y[test_index].values)
rf_mae, rf_rmse, rf_mape
```

```python
xgb = XGBRegressor(n_estimators=100)
xgb.fit(X[train_index], y[train_index])
xgb_pred = xgb.predict(X[test_index])
xgb_mae, xgb_rmse, xgb_mape = score_(xgb_pred, y[test_index].values)
xgb_mae, xgb_rmse, xgb_mape
```

```python
plt.bar(['linear', 'rf', 'xgb'], [linear_mae, rf_mae, xgb_mae])
plt.ylabel('MAE')
```

# Tunning and cross validation

```python
from sklearn.model_selection import GridSearchCV
```

```python
linear_model = Pipeline([('select_variable', PCA()),
                         ('modelling', LinearRegression())])

params={'select_variable__n_components':[2,3,4],
        'modelling__normalize':[True, False],
        'modelling__fit_intercept':[True, False]
        }
```

```python
regressor_tune = GridSearchCV(estimator=linear_model, param_grid = params, scoring='neg_mean_absolute_error',cv=3, refit=True)
regressor_tune.fit(X[train_index], y[train_index])
```

```python
regressor_tune.best_estimator_
```

```python
linear_tune_pred = regressor_tune.predict(X[test_index])
linear_tune_mae, linear_tune_rmse, linear_tune_mape = score_(linear_tune_pred, y[test_index].values)
linear_tune_mae, linear_tune_rmse, linear_tune_mape
```

```python
from sklearn.svm import LinearSVR
```

```python
linear_model = Pipeline([('select_variable', PCA()),
                         ('modelling', LinearSVR())])

params={'select_variable__n_components':[2,3,4],
        'modelling__C':[0.0001, 0.001, 1, 100, 10000],
        }


regressor_tune = GridSearchCV(estimator=linear_model, param_grid = params, scoring='neg_mean_absolute_error',cv=3, refit=True)
regressor_tune.fit(X[train_index], y[train_index])
```

```python
linear_tune_pred = regressor_tune.predict(X[test_index])
linear_tune_mae, linear_tune_rmse, linear_tune_mape = score_(linear_tune_pred, y[test_index].values)
linear_tune_mae, linear_tune_rmse, linear_tune_mape
```
